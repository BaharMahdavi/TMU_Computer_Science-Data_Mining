```python
import pandas as pd 
X = pd.read_csv(r"C:\Bahar\M.Sc Data Mining\Machine Learning (Oct 2022-Feb 2023)\HW1\Data\ORL10P_data.csv", header=None)
X
```

```python
import pandas as pd 
y = pd.read_csv(r"C:\Bahar\M.Sc Data Mining\Machine Learning (Oct 2022-Feb 2023)\HW1\Data\ORL10P_target.csv", header=None)
y
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2) 
```

```python
import numpy as np
from sklearn.linear_model import OrthogonalMatchingPursuit
```

```python
reg = OrthogonalMatchingPursuit(n_nonzero_coefs=30) 
reg.fit(X_train, y_train)
reg.score(X_train, y_train)
```

```python
reg.get_params(deep=True)
```

```python
reg.coef_
```

```python
reg.coef_.shape
```

```python
y_pred = reg.predict(X_test)
y_pred
```

```python
y_test
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error
MSE = mean_squared_error(y_test,y_pred)
print("Mean squared error Test=", round(mean_squared_error(y_test, y_pred), 2)) 
print('method is Linear regression by %s accuracy for test data'%((1-MSE)*100))
```
