```python
import pandas as pd 
X = pd.read_csv(r"C:\Bahar\M.Sc Data Mining\Machine Learning (Oct 2022-Feb 2023)\HW1\Data\CLL-SUB-111_data.csv", header=None)
X
```

```python
import pandas as pd 
y = pd.read_csv(r"C:\Bahar\M.Sc Data Mining\Machine Learning (Oct 2022-Feb 2023)\HW1\Data\CLL-SUB-111_target.csv", header=None)
y
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2) 
```

```python
X_train
```

```python
X_test
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X_train, y_train)
```

```python
reg.score(X_train, y_train)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
y_pred = reg.predict(X_train)
```

```python
import numpy as np
import sklearn.metrics as sm
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
MSE1 = sm.mean_squared_error(y_train, y_pred)
MAE1 = sm.mean_absolute_error(y_train, y_pred)
```

```python
print("Mean absolute error Train =", round(mean_absolute_error(y_train, y_pred), 2))
print("Mean squared error Train =", round(mean_squared_error(y_train, y_pred), 2))
```

```python
y_test_pred = reg.predict(X_test)
MSE2 = round(sm.mean_squared_error(y_test, y_test_pred),2)
MAE2 = round(sm.mean_absolute_error(y_test, y_test_pred),2)
```

```python
print("Mean absolute error Test=", round(sm.mean_absolute_error(y_test, y_test_pred), 2)) 
print("Mean squared error Test=", round(sm.mean_squared_error(y_test, y_test_pred), 2)) 
print('method is Linear regression  by %s  accuracy for test data'%((1-MSE2)*100))
print('method is Linear regression  by %s  accuracy for test data'%((1-MAE2)*100))
```

```python
import numpy as np
from sklearn import linear_model
reg = linear_model.Lasso (alpha = 0.5)
reg.fit(X_train, y_train)
```

```python
y_pred = reg.predict(X_test)
```

```python
y_pred
```

```python
import math
L = []
for i in y_pred:
    L.append((math.floor(i))) 
y_pred_lasso = np.expand_dims(L, -1)
```

```python
from sklearn.metrics import accuracy_score
Ac_Sc_L = accuracy_score(y_test, y_pred_lasso)
```

```python
print('method is Lasso by %s  accuracy for test data'%((1-Ac_Sc_L)*100))
```
